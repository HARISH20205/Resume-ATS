```python
import pandas as pd
import numpy as np
from datasets import load_dataset

```

```python
data = load_dataset("batuhanmtl/job-skill-set")
```

```python
data
```

```python
data = pd.DataFrame(data["train"])
```

```python
data.head()
```

```python
data.drop(columns=["job_id"],inplace=True)
```

```python
data
```

```python
data.drop_duplicates(inplace=True)
```

```python
data.fillna('', inplace=True) 
```

```python
import re
data.columns = data.columns.str.strip().str.lower().str.replace(' ', '_')
```

```python
data['job_description'] = data['job_description'].apply(lambda x: re.sub(r'\s+', ' ', x.strip().lower()))
data['job_skill_set'] = data['job_skill_set'].apply(lambda x: re.sub(r"['\[\]]", '', x.strip().lower()))
```

```python
data['job_skill_set'] = data['job_skill_set'].apply(lambda x: x.split(', '))
```

```python
data['category'] = data['category'].str.strip().str.upper()
```

```python
data['job_title'] = data['job_title'].apply(lambda x: re.sub(r'\(.*?\)', '', x).strip())
```
